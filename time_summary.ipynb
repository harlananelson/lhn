```python
# There isn't a built-in Python function that does exactly what you're asking for, but you can create one using the `os` and `datetime` modules. Here's a simple function that creates a timeline for when files in a directory were last modified:


import os
import datetime
from .header import get_logger

logger = get_logger(__name__)


def create_timeline(directory):
    timeline = {}
    for filename in os.listdir(directory):
        filepath = os.path.join(directory, filename)
        if os.path.isfile(filepath):
            timestamp = os.path.getmtime(filepath)
            modification_time = datetime.datetime.fromtimestamp(timestamp)
            timeline[filename] = modification_time
    return timeline

directory = '/path/to/your/directory'  # replace with your actual directory
timeline = create_timeline(directory)
for filename, modification_time in sorted(timeline.items(), key=lambda item: item[1]):
    print(f'{filename}: {modification_time}')


# In this code, `os.listdir(directory)` gets the names of the files in the directory. `os.path.join(directory, filename)` gets the full path of each file. `os.path.isfile(filepath)` checks if the path is a file. `os.path.getmtime(filepath)` gets the last modification time of the file. `datetime.datetime.fromtimestamp(timestamp)` converts the timestamp to a datetime object. `timeline[filename] = modification_time` adds the filename and its modification time to the timeline.

# `sorted(timeline.items(), key=lambda item: item[1])` sorts the items in the timeline by modification time. `print(f'{filename}: {modification_time}')` prints the filename and its modification time.

# Please replace `'/path/to/your/directory'` with the actual path to your directory. This function doesn't handle subdirectories. If you want to include subdirectories, you should modify the function to do so.
```
